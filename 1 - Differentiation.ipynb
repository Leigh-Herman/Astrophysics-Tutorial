Function to estimate the derivative of $ y(x) = \sin(x)$ over 10 equally spaced points.

Using this discrete data we will estimate the derivative at the third pont and the error with respect to the analytical value.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#First to write a function to approximate the evaluate the derivative at a given index idx

def left_side_deriv(x, f, idx):
   """return the left-sided derivative x[idx]"""
   return (f[idx] - f[idx-1])/(x[idx]-x[idx-1])

def right_side_deriv(x, f, idx):
   """return the right-sided derivative x[idx]"""
   return (f[idx+1] - f[idx])/(x[idx+1]-x[idx])

def centre_deriv(x, f, idx):
   """return the centre derivative of x[idx]"""
   return (f[idx+1] - f[idx-1])/(x[idx+1]-x[idx-1])
```

```python
#use x[ival] for the location of the derivative

ival = 3
```

```python
#create funciton and number of data points

x = np.linspace(0, np.pi, 10, endpoint=False)

f = np.sin(x)
```

```python
#calculate the derivatives

dl = left_side_deriv(x, f, ival)
dr = right_side_deriv(x, f, ival)
dc = centre_deriv(x, f, ival)
```

```python
#print results

print(f"left = {dl}, right = {dr}, centered = {dc}")
print(f"analytic value = {np.cos(x[ival])}")
```
